# Interactive Phase Portrait Explorer
Explore the phase portraits of planar linear systems $X' = AX$ in canonical form with eigenvalues $\lambda_1$ and $\lambda_2$.

Drag the sliders to change the eigenvalues and watch the phase portrait update.

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
from ipywidgets import interact, FloatSlider

angles1 = np.linspace(0, 2*np.pi, 9)[:-1]
angles2 = np.deg2rad([45, 135, 225, 315])

starts = [(np.cos(t), np.sin(t)) for t in angles1] + \
         [(2*np.cos(t), 2*np.sin(t)) for t in angles2] + \
         [(3*np.cos(t), 3*np.sin(t)) for t in angles2]

step = 0.02
n_steps = 300

def draw(lam1=-1.0, lam2=1.0):
    fig, ax = plt.subplots(figsize=(6, 6))
    for (x0, y0) in starts:
        fwd_x = [x0]
        fwd_y = [y0]
        for _ in range(n_steps):
            dx = lam1 * fwd_x[-1]
            dy = lam2 * fwd_y[-1]
            fwd_x.append(fwd_x[-1] + dx * step)
            fwd_y.append(fwd_y[-1] + dy * step)

        bwd_x = [x0]
        bwd_y = [y0]
        for _ in range(n_steps):
            dx = lam1 * bwd_x[-1]
            dy = lam2 * bwd_y[-1]
            bwd_x.append(bwd_x[-1] - dx * step)
            bwd_y.append(bwd_y[-1] - dy * step)

        ax.plot(fwd_x, fwd_y, 'b-')
        ax.plot(bwd_x, bwd_y, 'b-')

        dx = lam1 * x0
        dy = lam2 * y0
        speed = np.sqrt(dx**2 + dy**2)
        if speed > 0:
            ax.annotate('', xy=(x0 + dx/speed*0.2, y0 + dy/speed*0.2),
                        xytext=(x0, y0),
                        arrowprops=dict(arrowstyle='->', color='red', lw=1.5))

    ax.axhline(0, color='black', linewidth=0.5)
    ax.axvline(0, color='black', linewidth=0.5)
    ax.set_xlim(-5, 5)
    ax.set_ylim(-5, 5)
    ax.set_aspect('equal')
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_title(f'λ₁ = {lam1:.1f}, λ₂ = {lam2:.1f}')
    plt.show()

interact(draw,
         lam1=FloatSlider(min=-5, max=5, step=0.1, value=-1, description='λ₁'),
         lam2=FloatSlider(min=-5, max=5, step=0.1, value=1, description='λ₂'));
```
